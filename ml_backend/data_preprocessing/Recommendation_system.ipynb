```python
tfidf = TfidfVectorizer()
```

```python
df.sort_values(
    ['rating', 'rating_count'],
    ascending=False
)
```

```python
def recommend(title, brand, category, size):
    query= title + " " + brand + " " + category + " " + size
    query_vector= tfidf.transform([query])
    similarity= cosine_similarity(query_vector, tfidf_matrix)
    scores= similarity[0]
    valid_indices = np.where(scores > 0.20)[0]
    ratings = df['rating']
    normalized_rating = (ratings - ratings.min()) / (ratings.max() - ratings.min())
    final_score= 0.85 * (similarity[0]) + 0.15 * (normalized_rating)
    filtered_scores = final_score[valid_indices]
    
    sorted_indices = valid_indices[np.argsort(filtered_scores)[::-1]]
    top_indices = sorted_indices[:2]
    
    return df.iloc[top_indices][
    ['title', 'brand', 'category', 'current_price', 'rating', 'image_url']
    ]
```

```python
print (recommend('anarkali', 'h', 'kurtis', '28'))
```
